¿Porqué pueden faltar valores / datos en el dataset?

Principalmente hay 2 razones:

* La extracción de los datos: Puede ocurrir que haya incompatibilidades entre el tipo de datos entre la DB y el fichero de extracción.
* Recolección de los datos: En algunos casos son datos realmente inexistentes que nunca se llegaron a meter en la DB.

### Estrategias para lidiar con valores faltantes en la extracción de datos.

Existen varios enfoques con los que se puede actuar cuando se ha lidiar con datos faltantes:

* Drástico: Borramos el registro completo (la fila)
* Menos drástico: Borramos la columna (en el caso del ejemplo borrar "body")

```python
# Si tenemos un dataset en el que los valores 'null' suponen un 1%-3% de los registros puede ser buena idea borrar todas
# esas filas que tienen ese dato 'null'. Para ello lo mejor es usar la función .dropna(). Esta función tiene 2 parámetros muy
# importantes:
#
# · axis:    axis=0 --> Borra las filas que cumplan la condición dispuesta en el "how"
#            axis=1 --> Borra las columnas que cumplan la condición dispuesta en el "how"

# data.dropna(axis=0, how="all") --> Borra la fila del dataset solo si TODAS las columnas de dicha fila tienen valor 'null'
#                                    Este caso sería el más restrictivo y el que más protege contra grandes borrados.

# data.dropna(axis=0, how="any") --> Borra la fila del dataset si en alguna de sus columnas contiene un valor 'null'
```

```python
# Creo el set data_2 para realizarle un how="any". Veremos que data_2 al final queda completamente vacío.
# Esta solución es muy drástica :(
# Si bien, .dropna no realizar los cambios en el dataset como tal, es decir, no sobreescribe el dataset.
# Si queremos sobreescribir el data set tendríamos que poner la línea
# data_2 = data_2.dropna(axis=0, how="any")
data_2=data
data_2.dropna(axis=0, how="any")
```

# Inferencia de los datos

En contraposición al borrado de datos existe la posibilidad de la inferencia de datos para aquellos datos que estén faltantes en el set de datos.

Estos valores que se añaden pueden ser por ejemplo:

* Ceros '0'.
* Un cierto string
* El promedio del resto de valores no faltantes de esa columna

Existen varios métodos de imputación de datos faltantes.

```python
# Creamos un data_3 que se rellenará de ceros a los datos faltantes
# Como en el caso de .dropna, .fillna no reescribe a no ser que se le pase (con el ejemplo de data_3)
#
# data_3 = data_3.fillna(0)
data_3 = data
data_3.fillna(0)
data_3.head()
```

```python
# Ahora ya si reescribimos:
data_3 = data_3.fillna(0)
```

```python
data_3.head()
```

```python
# Este approach data_3.fillna(0) rellena tanto columnas de valor entero como columnas de valor string con 0s.
# Si queremos customizar el rellenado se hace de la siguiente manera (creamos data_4)
#

data_4 = data
data_4["body"] = data_4["body"].fillna(0)
data_4["home.dest"] = data_4["home.dest"].fillna("unknown")
data_4.head(24)
```

# Imputando datos promedio.

Se trata de imputar en los datos NaN el promedio de los datos de dicha columna que no son NaN

```python
# ¿Cuántos datos de age son Nan?
pd.isnull(data_4["age"]).values.sum()
```

```python
#Vamos a rellenar con el promedio
data_4["age"]=data_4["age"].fillna(data_4["age"].mean())
```

```python
# Se puede ver como los NaN se han reemplazado por 29.881135 :)
# Aparte del promedio, también se puede usar mediana o incluso ciertas funciones lambda de imputación de datos faltantes.
data_4["age"]
```

```python
# Vamos a usar ahora unos parámetros dentro de .fillna()
#
# *Forward Fill:  Reemplaza el valor faltante con el valor más cercano DESPUES del mimsmo.
# *Back Fill:     Reemplaza el valor faltante con el valor más cercano ANTES del mimsmo.
#
# Creo un data_6 para esto.
data_6 = pd.read_csv(fullpath)
```

# Creando Variables Dummy

Se trata de crear variables separadas para cada categoría si tenemos una variable categórica dentro del data set.

Ejemplo dentro del Titanic Set --> La variable "sex" que es una variable categórica con outputs Male / Female.
Lo que haremos es crear dos columnas: "Male" y "Female". La columna "male" tendrá 1s cuando el registro sea Male y 0s cuando el registro sea Female. Comportamiento opuesto se espera en la columna Female.

```python
# Esto que hemos hecho hasta ahora en dumificación tan "manual"... podemos hacerlo todo de golpe. Para ello, podemos crear
# una función que lo haga
#
def createDummies(dataframe, var_name):
    dummy_var = pd.get_dummies(dataframe[var_name], prefix=var_name)
    dataframe = dataframe.drop(var_name, axis=1)
    dataframe = pd.concat([dataframe, dummy_var], axis=1)
    return dataframe
```

```python
# Vamos a usar esta función recién creada en el data set data_3
#
createDummies(data_3, "sex")
```
